```python
import pandas as pd
import matplotlib.pyplot as plt

# c d liu t file CSV (c header)
df = pd.read_csv("./Simulation/data/cqi_ideal_condition_bsr100000.csv")

# V biu  ng th hin s thay i CQI theo TTI cho tng UE
plt.figure(figsize=(15, 8))
for ue in df.columns:
    plt.plot(df.index, df[ue], label=ue, marker='o', linewidth=1)

plt.title('S thay i CQI ca tng UE theo thi gian (TTI)')
plt.xlabel('Thi gian (TTI)')
plt.ylabel('Ch s CQI')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

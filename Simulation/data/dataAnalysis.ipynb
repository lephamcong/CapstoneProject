```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file CSV (có header)
df = pd.read_csv("./Simulation/data/cqi_ideal_condition_bsr100000.csv")

# Vẽ biểu đồ đường thể hiện sự thay đổi CQI theo TTI cho từng UE
plt.figure(figsize=(15, 8))
for ue in df.columns:
    plt.plot(df.index, df[ue], label=ue, marker='o', linewidth=1)

plt.title('Sự thay đổi CQI của từng UE theo thời gian (TTI)')
plt.xlabel('Thời gian (TTI)')
plt.ylabel('Chỉ số CQI')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
